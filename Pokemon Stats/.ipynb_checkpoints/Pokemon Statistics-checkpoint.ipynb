# Pokemon Stats (PRÁCTICA)


## 1. Problem Definition

> Con la información del dataset, es posible encontrar cuales son las dos variables más significativas en la defición del tipo de pokemon?

## 2. Data

This data set includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed. It has been of great use when teaching statistics to kids. 

This are the raw attributes that are used for calculating how much damage an attack will do in the games. This dataset is about the pokemon games (NOT pokemon cards or Pokemon Go).

* The data for this table has been acquired from several different sites, including:

* http://www.pokemon.com/us/pokedex/
* http://pokemondb.net/pokedex
* http://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number

## 3. Evaluation

One question has been answered with this database: The type of a pokemon cannot be inferred only by it's Attack and Deffence. It would be worthy to **find which two variables can define the type of a pokemon**, if any. Two variables can be plotted in a 2D space, and used as an example for machine learning. This could mean the creation of a visual example any geeky Machine Learning class would love.

We could use some correlation matrix for example.

## 4. Features

The data as described by Myles O'Neill is:

* #: ID for each pokemon.
* Name: Name of each pokemon.
* Type 1: Each pokemon has a type, this determines weakness/resistance to attacks.
* Type 2: Some pokemon are dual type and have 2.
* Total: sum of all stats that come after this, a general guide to how strong a pokemon is.
* HP: hit points, or health, defines how much damage a pokemon can withstand before fainting.
* Attack: the base modifier for normal attacks (eg. Scratch, Punch).
* Defense: the base damage resistance against normal attacks.
* SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam).
* SP Def: the base damage resistance against special attacks.
* Speed: determines which pokemon attacks first each round.

## Modelling

### Split the data

Debido a que estamos frente a un problema Multi-Label y Multi-Class, vamos a segmentarlo como dos problemas Single-Label+Multi-Class, donde el segundo label "y2" dependerá del primer label "y1". 

Sin embargo, tenemos 3 opciones:

1. Modelar para Type 1 eliminando Type 1 y Type 2, luego para Type 2 modelamos pero tomando en cuenta Type 1 original. (ESTA)

2. Modelar para Type 1 eliminando Type 1 y Type 2, luego para Type 2 modelamos pero tomando en cuenta Type 1 predecido.

3. Modelar para Type 1 eliminando Type 1 y Type 2, de igual forma para Type 2 pero sin tomar en cuenta Type 1. (No parece correcto)

```python
# Primera modelación
X1 = df_tmp.drop(["Name", "Type 1", "Type 2"], axis=1)    
y1 = df_tmp["Type 1"]

# Segunda modelación
X2 = df_tmp.drop(["Name", "Type 2"], axis=1)             
y2 = df_tmp["Type 2"]
```

Primero, modelaremos para el primer label.

```python
# Data split (Train & Test)
np.random.seed(42)
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)
```

```python
X1_train
```

```python
# Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "BaggingClassifier": BaggingClassifier(),
          "Random Forest": RandomForestClassifier(),
          "ExtraTreesClassifier": ExtraTreesClassifier(),
          "AdaBoostClassifier": AdaBoostClassifier(),
          "GradientBoostingClassifier": GradientBoostingClassifier(),
          "HistGradientBoostingClassifier": HistGradientBoostingClassifier()}

models
```

```python
#Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    '''
    Fits and evaluates given ML models.
    models: a dict of different Scikit-Learn ML models.
    X_train: training data (no labels)
    X_test: testing data (no labels)
    y_train: training labels
    y_test: test labels
    '''
    # Set Random Seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    
    # Loop trough models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
        
    return model_scores
```

**Type 1:**

```python
fit_and_score(models, X1_train, X1_test, y1_train, y1_test)
```

**Type 2:**

```python
fit_and_score(models, X2_train, X2_test, y2_train, y2_test)
```

Debido a que se trata de un problema Multi-Class, los modelos habituales no son para nada óptimos. Por ello, utilizaremos métodos multiclass, tales como OneVsOne, OneVsRest, etc.

```python
#Create a second function to fit and score models
def fit_and_score_multiclass(models, X_train, X_test, y_train, y_test):
    '''
    Fits and evaluates given ML models.
    models: a dict of different Scikit-Learn ML models.
    X_train: training data (no labels)
    X_test: testing data (no labels)
    y_train: training labels
    y_test: test labels
    '''
    # Set Random Seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    
    # Loop para OneVsRest
    for name, model in models.items():
        # Fit and score OneVsRest with the model
        clf = OneVsRestClassifier(model).fit(X_train, y_train)
        model_scores["OneVsRest "+name] = clf.score(X_test, y_test)
        
    # Loop para OneVsOne
    for name, model in models.items():
        # Fit and score OneVsRest with the model
        clf = OneVsOneClassifier(model).fit(X_train, y_train)
        model_scores["OneVsOne "+name] = clf.score(X_test, y_test)
        
    return model_scores
```

```python
fit_and_score_multiclass(models, X1_train, X1_test, y1_train, y1_test)
```

```python
fit_and_score_multiclass(models, X2_train, X2_test, y2_train, y2_test)
```

Con esto podemos concluir que la definición del Type 1 de cada pokemon, no tiene suficiente relación con el resto de estadísticas, sin embargo, el Type 2 si tiene una relación ligera según el Type 1.

```python
# Helper function for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({"features": columns,
                        "feature_importances": importances})
         .sort_values("feature_importances", ascending=False)
         .reset_index(drop=True))
    # Plot the dataframe
    fig, ax = plt.subplots()
    ax.barh(df["features"][:n], df["feature_importances"][:20])
    ax.set_ylabel("Features")
    ax.set_xlabel("Feature Importance")   
    ax.invert_yaxis()    
```
